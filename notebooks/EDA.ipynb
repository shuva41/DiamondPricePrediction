```python
#The columns 'cut', 'colour' and 'clarity' are ranked --> They are ordinal
df['cut'].unique()
```

```python
# Here Ideal is rank 1, Premium is rank 2, Very Good is rank 3, Good is rank 4, Fair is Rank 5 ----> Here you need the domain knowledge
```

```python
cut_map = {'Fair':1,'Good':2,'Very Good':3, 'Premium':4, 'Ideal':5}
```

```python
df['clarity'].unique()
```

```python
clarity_map = {"I1":1, "SI2":2,"SI1":3,"VS2":4,"VS1":5,"VVS2":6,"VVS1":7,"IF":8}
```

```python
df['color'].unique()
```

```python
color_map = {"D":1,"E":2,"F":3,"G":4,"H":5,"I":6,"J":7}
```

```python
# Replace the ordinal features with their corresponding ranks
df['cut'] = df['cut'].map(cut_map)
df['clarity'] = df['clarity'].replace(clarity_map)
df['color'] = df['color'].replace(color_map)
```
